### K Means Clustering Algorithm 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def k_means_clustering(data, K, max_iterations=1000, tol = pow(15,-2)): 
  N = data.shape[0]
  d = data.shape[1]

  labels = np.zeros(N)
  SSE_history = []
  
  centroids = np.random.rand(K,d)
  it = 0
  for i in range(max_iter):
    SSE_history.append(0)
    p = 0
    for x in range(len(data)):
      temp = np.empty(K)
      for j in range(K):
        temp[j] = np.linalg.norm(data[x]-centroids[j])
      labels[p] = np.argmin(temp)
      p+=1
    for k in range(K):
      t = []
      for s in range(len(data)):
        if (labels[s]==k):
          t.append(np.linalg.norm(data[s]-centroids[k]))
      sm = sum(t)
      SSE_history[it] = SSE_history[it] + sm
    for c in range(K):
      temp = []
      for j in range(len(labels)):
        if (labels[j] == c):
          temp.append(data[j])
    for c in range(K):
      if (data[labels==c].size == 0):
        continue
    if (len(SSE_history) > 1):
      if (np.absolute(SSE_history[it]-SSE_history[it-1])/SSE_history[it-1]<=tol):
        break
    it+=1

  return labels, centroids, SSE_history, iter
```
